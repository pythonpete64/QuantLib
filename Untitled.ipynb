```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from util import get_data, plot_data, compute_daily_returns
```

```python
dates = pd.date_range('2017-06-01', '2018-06-14')
symbols = ['BTCUSD', 'ETHUSD']
df = get_data(symbols, dates)
df = pd.DataFrame.dropna(df)
print(df)
df.info()
```

```python
# compute_daily_returns
dr = compute_daily_returns(df)

#plot scatter plots for all VS BTCUSD
dr.plot(kind='scatter', x='BTCUSD', y='ETHUSD')
beta_ETHUSD, alpha_ETHUSD = np.polyfit(dr['BTCUSD'], dr['ETHUSD'], 1)
plt.plot(dr['BTCUSD'], (dr['BTCUSD']*beta_ETHUSD) + alpha_ETHUSD, '-', color='r')
plt.show()
```
